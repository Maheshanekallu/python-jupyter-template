### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Data
x = list(range(0, 11))               # x values from 0 to 10
y = [i**2 for i in x]                # y values as square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='blue')

# Customize the plot
plt.title("Line Plot of x vs x²")       # Title
plt.xlabel("x values")                  # X-axis label
plt.ylabel("y = x²")                    # Y-axis label
plt.grid(True)                          # Enable grid

# Show the plot
plt.show()
```
